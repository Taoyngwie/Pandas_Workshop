```python
#ตอนที่ 27 - สอบถามข้อมูลแบบมีเงื่อนไข

import pandas as pd
```

```python
#เอา indexColumns มาไว้เป็นลำดับแรก
df = pd.read_excel("D:\Destop\Pandas_workshop\pandastutorial-main\dataupdate.xlsx","weather",index_col="Day") #แสดงหัว index ให้เป็นแบบวันที่ 
```

```python
#แสดงผลโดยที่เราใช้ Loop
for idx,row in df.iterrows():
    print("อุณหภูมิ = {} ส่งผให้เกิด {}".format(row.Temperature,row.Event))
```

```python
#แสดงวันทีมีอุณหภูมิ <= 20 องศา
df[df.Temperature<=20]
```

```python
#แสดงวันที่แดดร้อน

df[df.Event == "แดดร้อน"]
```

```python
#แสดงวันที่ฝนตก เเละ มากกว่าเท่ากับ 35

df.loc[(df.Event == "ฝนตก") & (df.Temperature >= 35)]
```

```python
#แสดงวันที่เมฆมาก หรือ อุณหภูมิ <= 18

df.loc[(df.Event == "เมฆมาก") | (df.Temperature <= 18)]
```

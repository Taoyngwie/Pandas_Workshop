```python
#การจัดการข้อมูลที่เป็นช่องว่าง
#1 เริ่มตรวจสอบโดยใช้ isnull()
#2 ตรวจสอบเข้าไปโดยใช้ .any()
```

```python
# มีคอลัมน์ใดบ้างที่มีค่าว่าง
df.isnull().any()
```

```python
#ตอนที่ 34 - การเพิ่มแถว

import pandas as pd
```

```python
#เอา indexColumns มาไว้เป็นลำดับแรก
df = pd.read_excel("D:\Destop\Pandas_workshop\pandastutorial-main\Stock.xlsx",index_col="Name") #แสดงหัว index ชื่อ
df
```

```python
#ทำการเพิ่มคอลัมน์
df["delivery"] = 100
```

```python
#ทำการเพิ่มคอลัมน์รวมราคาสินค้า โดยที่เอาคอลัมน์ที่เรากำหนดมาหาผลรวมกัน

df["total"] = df["Price"] + df["delivery"]
```

```python
#เรียง จำนวน จากน้อยไปมาก
result = df.sort_values("Amount",ascending=True)
result
```

```python
#ทำการเปลี่ยนชื่อคอลัมน์ แบบชื่ออันเดียว
# cols = {"ชื่อเดิม":"ชื่อใหม่"}

cols = {"Category":"Group"}
df.rename(columns=cols,inplace=True)
df
```

```python
#ทำการเปลี่ยนชื่อคอลัมน์ แบบชื่อหลายอัน
# cols = {"ชื่อเดิม":"ชื่อใหม่"}

cols = {"Price":"Cost","Amount":"Quantity","total":"Summation"}
df.rename(columns=cols,inplace=True)
df
```

```python
#ทำการลบคอลัมน์ delivery ออกไป เเบบถาวร
df.drop("delivery",axis=1,inplace=True)
df
```

```python
# ข้อมูลตัวใหม่ที่เราจะเพิ่ม
product = [["หูฟัง","อุปกรณ์คอมพิวเตอร์",400,20,500],
["สายชาร์จ","อุปกรณ์คอมพิวเตอร์",1400,20,1500]]
cols = ["Name","Group","Cost","Quantity","Summation"]
```

```python
# นำจากที่เราจะเพิ่มเอามาใส่ใน dataframe
newdf = pd.DataFrame(data=product,columns=cols)
newdf
```

```python
newdf.set_index("Name",inplace=True) #ทำหัว คอลัมน์ให้เป็น Name
newdf
```

```python
#นำข้อมูลจากที่เราจเพิ่มมาต่อที่ข้อมูลเดิม
df = df.append(newdf)
df
```

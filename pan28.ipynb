```python
#ตอนที่ 30 - เรียงลำดับข้อมูลด้วย Value

import pandas as pd
```

```python
#เอา indexColumns มาไว้เป็นลำดับแรก
df = pd.read_excel("D:\Destop\Pandas_workshop\pandastutorial-main\Stock.xlsx",index_col="Name") #แสดงหัว index ชื่อ
df
```

```python
#เรียงราคาจากน้อยไปมาก
result = df.sort_values("Price")
result
```

```python
#เรียงราคาจากมากไปน้อย
result = df.sort_values("Price",ascending=False)
result
```

```python
#เรียงราคาจากมากไปน้อย แล้วเซฟเป็นฉบับจริง
df.sort_values("Price",ascending=False,inplace=True)
df
```

```python
#เรียง จำนวน จากมากไปน้อย
result = df.sort_values("Amount",ascending=False)
result
```
